```python
def scraper_nypost(csv_ny_pad):

  ny_post = pd.read_csv(csv_ny_pad)
  link = ny_post['Link'].dropna()
  link = link.unique()
  link = link.tolist()
  df_notas_nypost = pd.DataFrame(columns = ['Titulo' , 'Url', 'Fecha', 'Texto'])

  for i in range(len(link)):

    print(i)
    try:

      time.sleep(1 * np.random.random())
      response = requests.get(link[i])
      soup = BS(response.content)
      titulo = soup.find(attrs = {"class": "headline headline--single"})
      titulo = re.sub(r'\t', '', titulo.text).strip()
      fecha = soup.find(attrs = {"class": "date meta meta--byline"})
      fecha = fecha.find('span')
      fecha = timestring.Date(fecha.text)
      cuerpo = soup.find(attrs = {"class": "single__content entry-content m-bottom "})
      cuerpo = cuerpo.find_all('p')
      texts = []
      for p in cuerpo:
        
        texts.append(p.text)

      body_text = '\n'.join(texts)
      df_notas_nypost = df_notas_nypost.append({'Titulo': titulo, 'Url': link[i] , 'Fecha': fecha, 'Texto': body_text}, ignore_index=True)

    except:
      
      pass

  return df_notas_nypost
```

```python
ny_post = scraper_nypost('/content/gdrive/MyDrive/Cendie/NY-Post.csv')
```
